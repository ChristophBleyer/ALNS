# Conclusions

With a great variance in travel times due to the nature of the routing environment the estimated average travetime is too high, such that even nearby customers can not be assigned anymore even though (as tested many times running the alns algorithm) can be included in a solution easily. The median performs better both in percent of customers assigned and in the resulting ojective function of the initially constructed route.
